```python
def read_first_line(file_path):
    file_path_txt= os.path.join(file_path,"groundtruth_rect.txt")
    with open(file_path_txt, 'r') as file:
        line = file.readline().strip()
    id,x,y,width,height,visible = list(map(int,line.split(',')))
    
    file_path_img = os.path.join(file_path,"img","00001.jpg")
    
    full_image = cv2.imread(file_path_img)
    template = full_image[y:y+height, x:x+width]
    return template
```

```python
region = read_first_line('datasets/MotorcycleChase')
plt.imshow(region)
plt.show()
```
